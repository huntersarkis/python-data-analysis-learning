## Jupyter Notebook

Jupyter Notebook is an interactive environment for writing and running Python code. It also allows you to include Markdown notes, visualizations, and output in one place.

### Key points:
- Start Jupyter with `jupyter notebook`
- Open the generated link in your browser
- New notebooks are saved as `.ipynb` files
- Run a cell with **Shift + Enter**
- Save with **File → Save and Checkpoint**
- Close properly with **File → Close and Halt**

## Helpful IPython and Jupyter Features

### Tab Completion
Pressing **Tab** helps autocomplete:
- variables
- methods
- modules
- file paths
- function arguments

This makes coding faster and reduces mistakes.

### Introspection
Using `?` before or after an object shows information about it.

Examples:
- `b?`
- `print?`

This can display:
- the object type
- function signature
- docstring
- other useful details

## Python Language Basics

Python is known for readability, simplicity, and explicit syntax.

### Indentation
Python uses indentation instead of braces to define code blocks.

Example:
- `if` statements
- `for` loops
- functions

Using consistent indentation is required. Four spaces is the recommended standard.

## Objects, Comments, and Functions

### Everything is an Object
In Python, values such as numbers, strings, lists, and functions are all objects. Each object has:
- a type
- data
- attributes
- methods

### Comments
Comments begin with `#` and are ignored by Python.
They are used to explain code or temporarily disable lines.

### Functions and Methods
Functions are called with parentheses.
Methods are functions attached to objects.

Examples:
- function: `print()`
- method: `my_string.upper()`

## Variables and References

In Python, variables store references to objects rather than copies of objects.

Example idea:
If `a` and `b` refer to the same list, changing one changes the other.

This is especially important when working with mutable data structures like lists and larger datasets.

## Dynamic References and Strong Typing

### Dynamic Typing
A variable can point to different types of objects at different times.

Example idea:
A variable can first hold an integer and later hold a string.

### Strong Typing
Python does not automatically combine incompatible types.

For example, adding a string and an integer directly causes an error.

Useful tools:
- `type()`
- `isinstance()`

## Attributes, Methods, and Duck Typing

### Attributes and Methods
Objects can have:
- attributes → stored data
- methods → functions attached to the object

These are accessed with dot notation.

### Duck Typing
Python often focuses on whether an object behaves a certain way rather than what exact type it is.

For example, an object can be treated as iterable if it can be looped over.

## Imports and Operators

### Imports
Modules are Python files containing reusable code.

Common import styles:
- `import module_name`
- `from module_name import something`
- `import module_name as alias`

### Operators
Python includes standard operators for:
- arithmetic: `+`, `-`, `*`, `/`, `//`, `**`
- comparisons: `==`, `!=`, `<`, `>`, `<=`, `>=`
- identity: `is`, `is not`

### Important difference
- `==` checks whether values are equal
- `is` checks whether two variables reference the same object

## Mutable vs Immutable Objects

### Mutable objects
These can be changed after creation:
- lists
- dictionaries
- NumPy arrays

### Immutable objects
These cannot be changed after creation:
- strings
- tuples

This matters because mutable objects can be modified in place, while immutable objects require creating a new object.

## Scalar Types

Python includes several built-in scalar types:

- `int` → whole numbers
- `float` → decimals
- `bool` → True or False
- `str` → text
- `bytes` → raw binary data
- `None` → null value

These are the basic building blocks used throughout Python programs.

## Numbers and Strings

### Numbers
Python mainly uses:
- `int`
- `float`

Division with `/` returns a float.
Floor division with `//` drops the remainder.

### Strings
Strings can use single or double quotes.
They are immutable, meaning they cannot be changed directly.

Important string concepts:
- slicing
- concatenation
- replacing text
- raw strings with `r""`
- formatting with `.format()` or f-strings

## Bytes, Unicode, Booleans, and Type Casting

### Bytes and Unicode
Python 3 uses Unicode strings by default.
Strings can be encoded into bytes and decoded back into text.

### Booleans
The two Boolean values are:
- `True`
- `False`

Boolean logic uses:
- `and`
- `or`
- `not`

### Type Casting
Values can be converted using:
- `str()`
- `int()`
- `float()`
- `bool()`

## None and Dates/Times

### None
`None` represents a null or missing value in Python.
It is commonly used as a default function argument or placeholder.

### Dates and Times
The `datetime` module provides tools for working with:
- dates
- times
- datetimes

Useful tasks include:
- formatting dates with `strftime()`
- parsing strings into dates with `strptime()`
- calculating differences using `timedelta`

## Control Flow

Python includes standard control flow tools:

### Conditional statements
- `if`
- `elif`
- `else`

These allow your code to make decisions.

### Loops
- `for` loops iterate over collections
- `while` loops repeat while a condition is true

### Other useful keywords
- `continue` → skip to the next iteration
- `break` → exit a loop
- `pass` → do nothing placeholder

## The `range()` Function

`range()` generates a sequence of integers and is commonly used in loops.

Examples of use:
- looping a certain number of times
- iterating through indexes
- creating start, stop, and step patterns

It is memory efficient because it generates values as needed.

## Tuples

A tuple is an ordered, fixed-length, and immutable sequence.

Examples:
- `(4, 5, 6)`
- `4, 5, 6`

### Key ideas:
- tuples cannot be changed after creation
- elements can be accessed with indexing
- tuples can contain mutable objects inside them
- tuples can be concatenated with `+`
- tuples can be repeated with `*`

### Unpacking
Tuples can be unpacked into variables.

Examples:
- `a, b, c = tup`
- `a, b = b, a`

You can also capture the remaining values with `*rest`.

## Lists

A list is an ordered, mutable, and variable-length sequence.

Lists are one of the most commonly used Python data structures.

### Common list methods:
- `append()` adds an element to the end
- `insert()` adds an element at a specific position
- `pop()` removes and returns an element
- `remove()` removes an element by value
- `extend()` adds multiple elements to a list
- `sort()` sorts the list in place

### Important ideas:
- lists can be modified in place
- `append()` is generally cheaper than `insert()`
- `extend()` is usually better than repeatedly using `+`
- membership can be checked with `in`

## Slicing

Slicing is used to select part of a sequence.

Examples:
- `seq[1:5]`
- `seq[:5]`
- `seq[3:]`
- `seq[-4:]`

### Key ideas:
- the start index is included
- the stop index is not included
- negative indices count from the end
- a step can be used, such as `seq[::2]`
- `seq[::-1]` reverses a sequence

## Dictionaries

A dictionary stores key-value pairs.

Example:
- `{"a": "some value", "b": [1, 2, 3]}`

### Common dictionary tasks:
- add or update a value
- access a value by key
- remove a value with `del` or `pop()`
- check whether a key exists with `in`

### Useful methods:
- `keys()`
- `values()`
- `items()`
- `update()`
- `get()`
- `setdefault()`

### Important idea:
Dictionary keys usually must be immutable and hashable, such as:
- strings
- integers
- tuples

## Creating Dictionaries from Sequences

Dictionaries can be created from paired sequences.

A common pattern is using `zip()` to combine keys and values.

Example idea:
- pair one sequence of keys with one sequence of values
- convert the result into a dictionary using `dict()`

This is useful in data preparation and mapping tasks.

## Default Values in Dictionaries

When working with dictionaries, it is common to need a fallback value if a key does not exist.

Helpful tools:
- `get()` returns a default value instead of raising an error
- `setdefault()` initializes a default value if the key is missing
- `defaultdict` from `collections` automates this process

These are especially useful when grouping values into lists.

## Sets

A set is an unordered collection of unique elements.

Sets are useful for:
- removing duplicates
- checking membership quickly
- comparing groups of values

### Common operations:
- union
- intersection
- difference
- symmetric difference

### Helpful methods:
- `add()`
- `remove()`
- `clear()`
- `issubset()`
- `issuperset()`
- `isdisjoint()`

Like dictionary keys, set elements generally must be immutable and hashable.

## Built-In Sequence Functions

Python includes several very useful sequence functions.

### `enumerate()`
Used when you want both the index and the value while looping.

### `sorted()`
Returns a new sorted list from any sequence.

### `zip()`
Pairs together elements from multiple sequences.

### `reversed()`
Iterates through a sequence in reverse order.

These are used often in data analysis and general Python programming.

## List, Set, and Dictionary Comprehensions

Comprehensions provide a concise way to create new collections.

### List comprehension
General form:
`[expression for value in collection if condition]`

Useful for:
- filtering values
- transforming values
- writing cleaner code

### Set comprehension
Uses curly braces and creates a set.

### Dictionary comprehension
Creates key-value mappings in one expression.

These are very common in Python and make many operations shorter and easier to read.

## Nested Comprehensions

Nested comprehensions can work with collections inside other collections.

They are useful for:
- flattening nested lists
- filtering values from nested structures

They are powerful, but too much nesting can make code harder to read.

## Functions

Functions are one of the main ways to organize and reuse code in Python.

They are created with the `def` keyword.

### Key ideas:
- functions can accept positional arguments
- functions can accept keyword arguments
- keyword arguments can have default values
- `return` sends a value back
- if no `return` is written, Python returns `None`

Functions improve readability, reuse, and structure.

## Scope and Namespaces

Variables created inside a function belong to the function’s local scope.

### Key ideas:
- local variables exist only while the function runs
- variables outside the function belong to a broader scope
- `global` can be used to modify global variables, but it is generally discouraged

Understanding scope helps prevent bugs and confusion in larger programs.

## Returning Multiple Values

Python can return multiple values from a function.

In practice, Python is returning a tuple, which can then be unpacked.

This is a very common and useful feature in Python.

## Functions as Objects

In Python, functions are objects.

That means functions can:
- be stored in variables
- be passed into other functions
- be reused in flexible ways

This is especially useful in data cleaning and transformation workflows.

## Lambda Functions

A lambda function is a short anonymous function written in one line.

They are useful when:
- passing a simple function into another function
- sorting with a custom rule
- applying quick transformations

Lambda functions are common in Python, especially in data work.

## Generators

A generator produces values one at a time instead of creating the full collection at once.

Generators use `yield` instead of `return`.

### Why generators matter:
- they are memory efficient
- they are useful for large data
- they work naturally in loops

Generators are important when working with large sequences or file data.

## Generator Expressions

Generator expressions are like list comprehensions, but they use parentheses instead of brackets.

They are useful when:
- passing values into functions like `sum()`
- reducing memory usage
- creating iterables lazily

They are often more efficient than building a full list first.

## itertools

The `itertools` module contains useful tools for working with iterables.

Examples include:
- `chain()`
- `combinations()`
- `permutations()`
- `groupby()`
- `product()`

These tools are useful for advanced looping and sequence operations.

## Errors and Exception Handling

Python code can raise exceptions such as:
- `ValueError`
- `TypeError`

To handle errors safely, Python uses:
- `try`
- `except`
- `else`
- `finally`

### Key ideas:
- `try` contains code that may fail
- `except` handles the error
- `else` runs if no error occurs
- `finally` always runs

Exception handling is important for building robust programs, especially when cleaning messy real-world data.

## Files and the Operating System

Python can open, read, write, and manage files using the built-in `open()` function.

### Common file modes:
- `"r"` = read
- `"w"` = write and overwrite
- `"x"` = create new file and fail if it exists
- `"a"` = append
- `"b"` = binary mode
- `"t"` = text mode

### Best practice:
Use `with open(...) as f:` so the file closes automatically when finished.

## Reading and Writing Files

### Common reading methods:
- `read()`
- `readlines()`
- loop through lines directly

### Common writing methods:
- `write()`
- `writelines()`

### Other useful file methods:
- `close()`
- `tell()`
- `seek()`

These methods are important for basic file handling and text processing.

## Bytes and Unicode with Files

Text mode works with strings, while binary mode works with bytes.

### Important ideas:
- UTF-8 is a common and recommended encoding
- it is best to specify `encoding="utf-8"` when opening text files
- binary mode reads raw bytes
- decoding problems can happen if byte sequences are incomplete
- `seek()` in text mode can be tricky when working with Unicode characters

This matters when dealing with non-ASCII text and real-world files.
